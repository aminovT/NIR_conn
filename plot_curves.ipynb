```python
s = '2.0_eval_results/lin_con_trans/LinearOneLayer{0}/curve.npz'
```

```python
import numpy as np
```

```python
b = np.load(s.format(10))
print(b.files)
```

```python
for i in b.files:
    print(i , ': ' ,  b[i])
```

```python
import matplotlib.pyplot as plt
plt.plot(b['tr_err'] , alpha = 0.2, label = '10')
plt.legend()
```

```python
plt.figure(figsize= (10,10))
for i in np.logspace(1,4 , 50):
    if int(i) != 10000:
        b = np.load(s.format(int(i)))
        plt.plot(b['tr_err'] , alpha = 0.2 , label = int(i))
plt.title('train error lin_con_trans')
plt.legend()
```

```python
plt.figure(figsize= (10,10))
for i in np.logspace(1,4 , 50):
    if int(i) != 10000:
        b = np.load(s.format(int(i)))
        plt.scatter( int(i) , b['tr_err_max'] , alpha = 1, label = int(i))
plt.xscale('log')
plt.title('tr_err_max lin_con_trans')
plt.legend()
```
